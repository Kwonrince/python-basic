## 0. 클래스 생성 
- 빈 사각형, 꽉 찬 사각형 출력하는 클래스(Box) 구현
- 생성자 : 사각형 가로 길이(width), 사각형 세로 길이 (length)
- 메소드 : 꽉 찬 사각형(black_box), 빈 사각형(white_box)
- black_box, white_box의 출력 예시는 아래와 같음
- 세로 길이 3, 가로 길이 7 일 때 예시임

```python
class Box():
    def __init__(self, width, length):
        self.width = width
        self.length = length
    
    def black_box(self):
        for i in range(self.length):
            print("*"*self.width)
    
    def white_box(self):
        print("*"*self.width)
        for i in range(self.length-2):
            print("*"+" "*(self.width-2)+"*")
        print("*"*self.width)
```

```python
box = Box(7, 3)
```

```python
box.black_box()
```

```python
box.white_box()
```

## 8.  새로운 컬럼 추가 1
- 미세먼지 농도별 예보 등급을 기준으로, 
- pm10 농도 '좋음'에 해당하는 값들은 'good','보통'의 경우 'soso', '나쁨'의 경우 'bad', '매우 나쁨'의 경우 
'very bad'의 값을 가지는 컬럼(컬럼명 : 'new_pm10')을 새로 추가
- pm2.5 농도 '좋음'에 해당하는 값들은 'good','보통'의 경우 'soso', '나쁨'의 경우 'bad', '매우 나쁨'의 경우 
'very bad'의 값을 가지는 컬럼(컬럼명 : 'new_pm2.5')을 새로 추가

```python
df.loc[df['hour_bef_pm10'] >= 0, 'new_pm10'] = 'good'
df.loc[df['hour_bef_pm10'] >= 31, 'new_pm10'] = 'soso'
df.loc[df['hour_bef_pm10'] >= 81, 'new_pm10'] = 'bad'
df.loc[df['hour_bef_pm10'] >= 151, 'new_pm10'] = 'very bad'
```

```python
df.loc[df['hour_bef_pm2.5'] >= 0, 'new_pm2.5'] = 'good'
df.loc[df['hour_bef_pm2.5'] >= 16, 'new_pm2.5'] = 'soso'
df.loc[df['hour_bef_pm2.5'] >= 36, 'new_pm2.5'] = 'bad'
df.loc[df['hour_bef_pm2.5'] >= 76, 'new_pm2.5'] = 'very bad'
```

## 9. 그래프 그리기 1
- pm10 기준, 미세먼지 등급이 '좋음', '보통','나쁨','매우 나쁨' 일 때 따릉이 대여 수의 평균값을 수직 막대그래프로 출력



```python
plt.bar(x = df['new_pm10'].unique(), height=df.groupby(by='new_pm10').mean()['count'])
plt.show()
```

## 10. 그래프 그리기 2
- pm2.5 기준, 미세먼지 등급이 '좋음', '보통','나쁨','매우 나쁨' 일 때 따릉이 대여 수의 평균값을 수평 막대그래프로 출력
- 그래프 제목은 '미세먼지(pm10) 등급에 따른 평균 따릉이 대여 수'로 지정


## 11. 새로운 컬럼 추가 2
- 오존 컬럼의 값을 내림차순으로 정렬하여, 해당 값을 새로운 컬럼(컬럼명 : 'new_ozone')으로 추가

```python
df['new_ozone'] = np.array(df['hour_bef_ozone'].sort_values(ascending=False))
```
