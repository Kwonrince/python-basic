```python
import pandas as pd
import numpy as np
import time
import re
from selenium import webdriver as wd
from bs4 import BeautifulSoup
from datetime import datetime
import time
```

```python
driver = wd.Edge("C:/Users/Soonki Kwon/Desktop/msedgedriver.exe")
driver.implicitly_wait(3)
driver.get('https://patents.google.com/')
elem = driver.find_element_by_id("searchInput")
elem.send_keys('metabus')
elem.send_keys(wd.common.keys.Keys.RETURN)
driver.implicitly_wait(2)

all_title, all_num, all_country, all_inventor = [], [], [], []
for i in range(0, 25):
    driver.find_element_by_xpath(f'/html/body/search-app/search-results/search-ui/div/div/div/div/div/div[1]/section/search-result-item[{i%10+1}]/article/state-modifier/a/h3/raw-html/span').click()
    time.sleep(1)
    
    page = driver.page_source
    soup = BeautifulSoup(page, 'html.parser')
    titles = soup.find('h1', id='title').text
    title = re.sub('\n +', '', titles)
    num = soup.find('h2', id='pubnum').text
    country = soup.find(class_='tagline style-scope patent-result').text
    inventor = soup.find(class_='important-people style-scope patent-result')
    inventor = str(inventor).split('data-inventor="')
    temp_inventor = []
    for j in inventor[1:]:
        temp_inventor.append(j.split('">')[0])
    
    if len(temp_inventor) == 1:
        temp_inventor = temp_inventor[0]
    else:
        temp_inventor = ','.join(temp_inventor)
    
    all_title.append(title)
    all_num.append(num)
    all_country.append(country)
    all_inventor.append(temp_inventor)
    time.sleep(0.5)
    
    driver.back()
    time.sleep(0.5)
    
    if (i%10+1) == 10:
        driver.find_element_by_xpath('/html/body/search-app/search-results/search-ui/div/div/div[2]/div/div/div[1]/div[5]/search-paging/state-modifier[3]/a/paper-icon-button/iron-icon').click()
        time.sleep(1)

driver.close()
```

```python
patent = pd.DataFrame({'title':all_title,
                       'country':all_country,
                       'num':all_num,
                       'inventor':all_inventor})
```
